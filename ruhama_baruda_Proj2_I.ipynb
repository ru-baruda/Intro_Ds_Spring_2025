## Chosen Dataset

This dataset features over 900 classic hits from the 1990s, with more than 500 different artists. Each track is described by in-depth Spotify audio features data such as danceability, energy, key, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo, and duration. ### 5. Provide two examples of DataFrame filtering using logical conditions

#### Filtering by column labels

```python
#selecting columns: Track, Artist, Popularity, and Year
print("Filtered DataFrame showing tracks, artists, popularity, and year:\n")
df.loc[:, ["Track", "Artist", "Popularity", "Year"]]
```

#### Filtering by rows

```python
#selecting rows where year is after 1995
print("Songs released after 1995:\n")
yearfilter_df = df.loc[:, ["Track", "Artist", "Year"]]
yearfilter_df[yearfilter_df.Year > 1995]
```

```python
#selecting rows where danceability is above 0.5 (scale of 0-1.0)
print("Songs that score above 0.5 in danceability:\n")
#new df with relevant columns
dance_df = df.loc[:, ["Track", "Artist", "Danceability"]]
#select rows where danceability value is greater than 0.5
dance_dfsort = dance_df[dance_df.Danceability > 0.5]
#sort new df in ascending order
dance_dfsort.sort_values(by = "Danceability")
```

### 7. Use .value_counts() to analyze at least two categorical columns

```python
#new df with categorical columns only
cat_columns = df.select_dtypes(include=['category', 'object'])
#value counts for artist column only
cat_columns["Artist"].value_counts()
```

```python
#convert year to categorical data to count how many tracks each year
cat_year_df = df["Year"].astype('category')
#count how many tracks in each year
print("Count of tracks in each year:\n")
cat_year_df.value_counts()
```

```python
#value counts for track column only
print("Count of track titles:\n")
cat_columns["Track"].value_counts()
```

### 8. Use .groupby() to compute summary statistics across categories

```python
#new df with relevant columns
group_df = df.loc[:, ["Track", "Artist", "Popularity", "Danceability"]]

# group by Artist and find averages of popularity and danceability
artist_grouped = group_df.groupby("Artist")[["Popularity", "Danceability"]].mean()
print("Average popularity and danceability of each artist:\n")
artist_grouped
```
